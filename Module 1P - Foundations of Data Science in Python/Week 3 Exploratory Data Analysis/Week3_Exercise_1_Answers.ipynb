#### Import Libraries

```python
import numpy
import pandas

%matplotlib inline
import matplotlib.pyplot as plt
```

#### Ex.1.1 Using matplotlib to draw a exponential curve with the following requirement：
<ul>
    <li>Line Color : Green</li>
    <li>Line Style : Dashed line</li>
    <li>Axes Limit : x:[-2, 2], y:[0, 9]</li>
    <li>Labels for x axis (x) and y axis (exp)</li>
    <li>Legend of Figure</li>
    <li>Title : A Exponential Curve</li>
</ul>


**Hint** : search function of numpy for a exponential sequence

```python
# code for Ex.1.1

# Create a sequence of expenential values
x = numpy.linspace(-2, 2, 50)  # generate 50 numbers with fixed interval from -2 to 2
y = numpy.exp(x)

# Create Figure object
fig = plt.figure()

# Draw figure
plt.plot(x, y, linestyle='--', color='g')   # with dashed line

# Axis limit
plt.axis([-2, 2, 0, 9])

# Adding Title and axes labels
plt.title('A Exponential Curve')
plt.xlabel('x')
plt.ylabel('exp');

# Display legend based on "label" parameter defined above in plt.plot()
plt.legend(['exp(x)']);

# Display the Plot
plt.show()
```

#### Ex.1.2 Using matplotlib to draw a exponential curve with the following requirement：
<ul>
    <li>Draw the curve with Scatter plot</li>
    <li>Marker Color Value : related curve value</li>
    <li>Marker Color : with colormap viridis</li>
    <li>Marker Style : triangle</li>
    <li>Marker Size : 100 times of related curve value</li>
    <li>Transparency : 0.3</li>
    <li>Axes Limit : x:[-2.5, 2.5], y:[-1, 9]</li>
    <li>Labels for x axis (x) and y axis (exp)</li>
    <li>Title : A Exponential Curve</li>
    <li>Colorbar</li>
</ul>

```python
# code for Ex.1.2

# Create a sequence of expenential values
x = numpy.linspace(-2, 2, 40)  # generate 40 numbers with fixed interval from -2 to 2
y = numpy.exp(x)

# Create Figure object
fig = plt.figure()

# Draw figure
plt.scatter(x, y, marker='<', c=y, s=y*100, alpha=0.3, cmap='viridis')

# Axis limit
plt.axis([-2.5, 2.5, -1, 9])

# Adding Title and axes labels
plt.title('A Exponential Curve')
plt.xlabel('x')
plt.ylabel('exp');

# Colorbar
plt.colorbar();

# Display the Plot
plt.show()
```

#### Ex.1.3 Draw a histogram of 1000 randomly generated numbers following normal distribution with the requirements shown below:
<ul>
    <li>Bin Number : 50</li>
    <li>Transparency : 0.5</li>
    <li>Labels for x axis (bins) and y axis (frequency)</li>
    <li>Title : A Histogram</li>
</ul>

```python
# code for Ex.1.3

# Generate 1000 data with normal/gaussian distribution
data = numpy.random.randn(1000)


# Draw histogram
plt.hist(data, bins=50, alpha=0.5);

# Adding Title and axes labels
plt.title('A Histogram')
plt.xlabel('bins')
plt.ylabel('frequency');

# Display the Plot
plt.show()
```
