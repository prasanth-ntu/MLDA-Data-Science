#### Import Libraries & Dataset

```python
import numpy
import pandas
%matplotlib inline
import matplotlib.pyplot as plt

# import the marathon dataset
data = pandas.read_csv('marathon-data.csv')
data.head()

# transform format"hh:mm:ss" to seconds
def time_convert(x):
    h,m,s = map(int,x.split(':'))
    return (h*60+m)*60+s

data['split'] = data['split'].apply(time_convert)
data['final'] = data['final'].apply(time_convert)
```

#### Ex.2.1 Compare the number of participants of the marathon by using bar plot directly from Series.

```python
# code for Ex.2.1

# Count the number of male and female participants
genderCount = data.groupby('gender').count()

# draw results by bar plot
genderCount.iloc[:,0].plot.bar(alpha=0.5)
```

#### Ex.2.2 Compare the number of participants of the marathon by using pie plot directly from Series.

```python
# code for Ex.2.2

# Count the number of male and female participants
genderCount = data.groupby('gender').count()

# draw results by bar plot
axis = genderCount.iloc[:,0].plot.pie(colormap='Greens')
axis.set_aspect('equal')
```

#### Ex.2.3 Draw a histogram of the ages of participants

```python
# code for Ex.2.3

# get ages
age = data['age']

# draw histogram
age.hist(alpha=0.7, bins=30)
```

#### Ex.2.4 Draw histograms of the ages of male and female participants seperately

```python
# code for Ex.2.4

# get data
age_byGener = data.groupby('gender')['age']

# draw histograms
age_byGener.plot.hist(alpha=0.7)   # directly draw from groupBy
plt.legend()
```

#### Ex.2.5 Draw a scatter to show the relationship between average time of splits and final time.

```python
# code for Ex.2.5

data.plot.scatter(x='split', y='final', alpha=0.1)
```

#### Ex.2.6 Draw histograms of final time of male and female seperately

```python
# code for Ex.2.5

# get ages
ftime_byGener = data.groupby('gender')['final']

# draw histogram
ftime_byGener.plot.hist(alpha=0.7, bins=30)
plt.legend()
```
